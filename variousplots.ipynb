```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'distance':[2,4,6,8,10],'time':[1,2,3,4,5]})
```

```python
plt.plot(df)
plt.ylabel('distance(km)')
plt.title("Simple plot with multilple lines on same graph")
```

```python
distance = [2,4,6,8,10]
time = [1,2,3,4,5]
plt.scatter(distance,time)
plt.title("Scatter plot")
```

```python
df=pd.DataFrame({'distance':[2,4,6,8,10],'time':[1,2,3,4,5]})
plt.hist(df)
plt.xlabel('time(s)')
plt.ylabel('distance(km)')
plt.title("Histogram")
```

```python
distance = [2,4,6,8,10]
time = [1,2,3,4,5]
plt.plot(distance,time)
plt.xlabel('time(s)')
plt.ylabel('distance(km)')
plt.title("Simple plot")
```

```python
df=pd.DataFrame({'distance':[2,4,6,8,10],'time':[1,2,3,4,5]})
plt.plot(df, color='green', linestyle='dashed', marker='o',) 
plt.xlabel('time(s)')
plt.ylabel('distance(km)')
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.arange(0,40)
y=np.sin(x)
plt.plot(x, y)
plt.title("Curve")
plt.show()
```

```python
distance = [1, 2, 3, 4, 5] 
time = [10, 24, 36, 40, 5]  
tick_label = ['one', 'two', 'three', 'four', 'five']  
plt.bar(distance, time, tick_label = tick_label, 
        width = 0.8, color = ['red', 'green']) 
plt.title("Bar graph")
plt.show()
```
